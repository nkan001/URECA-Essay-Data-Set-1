# Using Word Count to predict Resolved Scores

```python
#Counting the number of words
```

```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
w_tokenized_text1 = word_tokenize(text1)
w_tokenized_text1
```

```python
len(w_tokenized_text1)
```

```python
#remove punctuations
def remove_punctuation_from_a_text(the_text):
    no_punctuation_tokenized_text1 = []
    count = 0
    for i in the_text:
        for e in [',', '.', '?', "'", '!', '-', '(', ')']:
            if i != e:
                count += 1
        if count == 8:
            no_punctuation_tokenized_text1.append(i)
        count = 0
    length_of_text = len (no_punctuation_tokenized_text1)
    return length_of_text
    
```

```python
remove_punctuation_from_a_text(w_tokenized_text1)
```

```python
# removing punctuation and doing word count for all texts
word_count = []
for element in range(0, 1783):
    essay = str(df.iloc[element, 2])
    #print(essay)
    w_tokenized_text = word_tokenize(essay)
    #print(w_tokenized_text)
    word_count.append(remove_punctuation_from_a_text(w_tokenized_text))
```

```python
#relationship btw word count and resolved grade
```

```python
df_word_count = pd.DataFrame(word_count)
wc_train = pd.DataFrame(df_word_count[:1426])
wc_test = pd.DataFrame(df_word_count[-357:])
```

```python
wc_train.describe()
```

```python
resolved_score = df['domain1_score']
df_resolved_score = pd.DataFrame(resolved_score)
rs_train = pd.DataFrame(df_resolved_score[:1426])
rs_test = pd.DataFrame(df_resolved_score[-357:])
```

```python
rs_train.describe()
```

```python
sb.jointplot(wc_train, rs_train)
```

```python
jointDF = pd.concat([rs_train, wc_train], axis = 1)
sb.heatmap(jointDF.corr(), vmin = -1, vmax = 1, annot = True)
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(wc_train, rs_train)
```

```python
print('Intercept \t: b = ', linreg.intercept_)
print('Coefficients \t: a = ', linreg.coef_)
```

```python
regline_x = wc_train
regline_y = linreg.intercept_ + linreg.coef_ * wc_train

f, axes = plt.subplots(1,1,figsize = (16,8))
plt.scatter(wc_train, rs_train)
plt.plot(regline_x, regline_y, 'r-', linewidth = 3)
plt.show()
```

```python
rs_train_pred = linreg.predict(wc_train)
rs_test_pred = linreg.predict(wc_test)
print("Explained Variance (R^2) \t:", linreg.score(wc_train, rs_train))
def mean_sq_error(actual, predicted):
    return np.mean(np.square(np.array(actual) - np.array(predicted)))
mse = mean_sq_error(rs_train, rs_train_pred)
print("Mean Squared Error (MSE) \t: ", mse)
print("Root Mean Squared Error (RMSE) \t:", np.sqrt(mse))
```

```python
# removing stop words
```

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words("english")) #stop words are all in lower case
stop_words.add('i')
```

```python
def remove_stopwords(text):
    w_lowercase = []
    for i in text:
        if i not in stop_words:
            w_lowercase.append(i.lower())
    return w_lowercase
```

```python
no_stopwords = remove_stopwords(w_tokenized_text1)
```

```python
no_stopwords_punc = remove_punctuation_from_a_text(no_stopwords)
```

```python
print("Length of original text: ", len(w_tokenized_text1))
print("Length withOUT stopwords AND without punctuation: ", no_stopwords_punc)
```
